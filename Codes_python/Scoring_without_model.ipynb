## Ciblage sur profilage

Dans cette section, nous discuterons de notre approche de ciblage aléatoire, ses objectifs et les performances qu'elle génère.

### Objectifs

Notre objectif principal avec le ciblage aléatoire est de :

- **Établir un niveau de référence :** Le ciblage aléatoire nous permet d'évaluer les performances de manière impartiale, en fournissant un point de comparaison pour nos autres méthodes de ciblage.
  
- **Analyser les performances :** Nous cherchons à comprendre comment notre système se comporte lorsqu'il cible aléatoirement les utilisateurs, afin d'identifier les zones d'amélioration potentielles et d'optimiser nos stratégies de ciblage.



```python

# Créer une liste vide pour stocker les données
donnees = []

# Ouvrir le fichier en mode lecture
with open("Data/base_telecom_2022_12.txt", "r") as fichier:
    # Lire les lignes du fichier
    lignes = fichier.readlines()

    # Parcourir chaque ligne
    for ligne in lignes:
        # Diviser la ligne en fonction du point-virgule et ajouter à la liste des données
        donnees.append(ligne.strip().split(";"))

# Créer un DataFrame à partir des données
df_profil= pd.DataFrame(donnees[1:], columns=donnees[0])  # Utiliser la première ligne comme en-tête

# Afficher le DataFrame
df_profil.head()

```

```python
# Listes des colonnes numériques
num_colonne=["duree_offre_init","duree_offre","nb_migrations","nb_services","nb_reengagements",
             "vol_appels_m6","vol_appels_m5","vol_appels_m4","vol_appels_m3","vol_appels_m2","vol_appels_m1",
             "vol_appels_m6","vol_appels_m5","nb_sms_m4","nb_sms_m3","nb_sms_m2","nb_sms_m1"]

# Listes des colonnes discrètes 
discrete_colonne=["sexe","csp","code_postal","type_ville","mode_paiement","flag_migration_hausse","flag_migration_baisse","flag_personnalisation_repondeur",
             "telephone_init","telephone","situation_impayes","flag_appels_vers_international","flag_appels_depuis_international","segment"]

```

```python
# Transformation en colonnes numériques
for i in num_colonne:
    df_profil[i] = df_profil[i].apply(pd.to_numeric)
```

```python
# Séparation de la base de données
df_resi=df_profil[df_profil["flag_resiliation"]=="1"]
df_non_resi=df_profil[df_profil["flag_resiliation"]=="0"]
```

```python

# Calculer le nombre de colonnes pour les sous-graphiques
num_cols = min(len(num_colonne), 3)  # Utiliser au maximum 3 colonnes

# Calculer le nombre de lignes pour les sous-graphiques
num_rows = -(-len(num_colonne) // num_cols)  # Arrondir vers le haut

# Créer une figure et des axes avec le nombre approprié de lignes et de colonnes
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 5*num_rows))

# Boucler sur les colonnes et les sous-graphiques
for i, colonne in enumerate(num_colonne):
    row = i // num_cols  # Calculer le numéro de ligne
    col = i % num_cols   # Calculer le numéro de colonne
    # Tracer la courbe de densité pour chaque colonne de la première base de données
    sns.kdeplot(data=df_resi[colonne], label=f'1 - Résilié - {colonne}', color='blue', shade=True, ax=axes[row, col])
    # Tracer la courbe de densité pour chaque colonne de la deuxième base de données
    sns.kdeplot(data=df_non_resi[colonne], label=f'0 - Non résilié - {colonne}', color='orange', shade=True, ax=axes[row, col])
    # Ajouter un titre pour chaque sous-graphique
    axes[row, col].set_title(f'Courbe de densité pour {colonne}')
    # Ajouter une légende pour chaque sous-graphique
    axes[row, col].legend()

# Ajouter un titre global à la figure
fig.suptitle('Courbes de densité des colonnes des deux bases de données')

# Ajuster l'espacement entre les sous-graphiques
plt.tight_layout()

# Afficher le graphique
plt.show()
```

```python

# Calculer le nombre de lignes nécessaire pour afficher les boxplots
num_rows = (len(num_colonne) + 2) // 3  # Round up division

# Créer une figure et des axes
fig, axes = plt.subplots(num_rows, 3, figsize=(15, 5*num_rows))

# Boucle sur chaque variable caractéristique pour tracer les boxplots
for i, variable in enumerate(num_colonne):
    row = i // 3
    col = i % 3
    sns.boxplot(data=df_profil, x='flag_resiliation', y=variable, ax=axes[row, col])
    axes[row, col].set_title(f'Boxplot de {variable} entre les deux groupes')

# Supprimer les axes non utilisés
for i in range(len(num_colonne), num_rows * 3):
    row = i // 3
    col = i % 3
    fig.delaxes(axes[row, col])

# Ajuster l'espacement entre les sous-graphiques
plt.tight_layout()

# Afficher la figure
plt.show()
```

```python

# Définir le nombre de colonnes et de lignes pour les sous-graphiques
num_cols = 3  # Nombre de colonnes pour les sous-graphiques
num_rows = -(-len(discrete_colonne) // num_cols)  # Arrondi supérieur de la division

# Créer une figure et des sous-graphiques
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5*num_rows))

# Tracer la distribution de chaque variable discrète entre les deux groupes
for i, variable in enumerate(discrete_colonne):
    ax = axes[i // num_cols, i % num_cols]  # Sélectionner le sous-graphique approprié
    sns.countplot(data=df_profil, x=variable, hue='flag_resiliation', ax=ax)
    ax.set_title(f'Distribution de {variable} entre les deux groupes')
    ax.set_xlabel('Catégories de la variable')
    ax.set_ylabel('Fréquence')

# Ajuster l'espacement entre les sous-graphiques
plt.tight_layout()

# Afficher le graphique
plt.show()
```

```python

# Utiliser la fonction crosstab de pandas pour créer la table de contingence
table_contingence = pd.crosstab(df_profil['segment'], df_profil['flag_resiliation'])


# Calculer les proportions conditionnelles par ligne
proportions_conditionnelles_ligne = table_contingence.div(table_contingence.sum(axis=1), axis=0)

# Afficher les proportions conditionnelles par ligne
print("\nProportions conditionnelles par ligne:")
print(proportions_conditionnelles_ligne)
```

```python
# Utiliser la fonction crosstab de pandas pour créer la table de contingence
table_contingence = pd.crosstab(df_profil['flag_migration_baisse'], df_profil['flag_resiliation'])

# Calculer les proportions conditionnelles par ligne
proportions_conditionnelles_ligne = table_contingence.div(table_contingence.sum(axis=1), axis=0)

# Afficher les proportions conditionnelles par ligne
print("\nProportions conditionnelles par ligne:")
print(proportions_conditionnelles_ligne)
```

```python
# Utiliser la fonction crosstab de pandas pour créer la table de contingence
table_contingence = pd.crosstab(df_profil['csp'], df_profil['flag_resiliation'])

# Calculer les proportions conditionnelles par ligne
proportions_conditionnelles_ligne = table_contingence.div(table_contingence.sum(axis=1), axis=0)

# Afficher les proportions conditionnelles par ligne
print("\nProportions conditionnelles par ligne:")
print(proportions_conditionnelles_ligne)
```

```python
from scipy.stats import chi2_contingency


# Créer une table de contingence à partir de votre DataFrame
table_contingence = pd.crosstab(df_profil['flag_resiliation'], df_profil['segment'])

# Effectuer le test du Chi carré
chi2, p_value, dof, expected = chi2_contingency(table_contingence)

# Afficher les résultats
print("Test du Chi carré :")
print(f"Statistique du test : {chi2}")
print(f"Valeur-p : {p_value}")
print(f"Degrés de liberté : {dof}")
```
